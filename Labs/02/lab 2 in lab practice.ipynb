```python
#Simple Reflex Agent Code Example #2 (Vacuum Cleaner)
class Environment: #environment class
    def __init__(self, state='Dirty'):
        self.state = state
    
    def get_percept(self): #get perception from environment
        return self.state
    
    def clean_room(self):
        self.state = 'Clean'

class SimpleReflexAgent: #agent class
    def __init__(self):
        pass
    
    def act(self, percept):
        if percept == 'Dirty':
            return 'Clean the room'
        else:
            return 'Room is already clean'
        
#simulation function
# -> simulates interactions between the agent and the environment for a specified num of steps
# -> prints the percept and action for each step

def run_agent(agent, environment, steps):
    for step in range(steps):
        percept = environment.get_percept()
        action = agent.act(percept)
        print(f"Step {step + 1}: Percept - {percept}, Action - {action}")
        
        if percept == 'Dirty':
            environment.clean_room()

#create instances of agent and environment
agent = SimpleReflexAgent()
environment = Environment()

#run the agent in the environment for 5 steps
run_agent(agent, environment, 5)
```

```python
class Environment:
    def __init__(self):
        #create a 3x3 grid
        self.grid = ['Clean', 'Dirty', 'Clean',
                     'Clean', 'Dirty', 'Clean',
                     'Clean', 'Clean', 'Clean']
    
    def get_percept(self, position):
        #return the state of the current position
        return self.grid[position]
    
    def clean_room(self, position):
        #clean the room at the given position
        self.grid[position] = 'Clean'
        
    def display_grid(self, agent_position):
        #display the current state of the grid
        print("\ncurrent grid state:")
        grid_with_agent = self.grid[:] #copy grid
        grid_with_agent[agent_position] = "游놓" #place the agent at the current position
        
        for i in range(0, 9, 3):
            print(" | ".join(grid_with_agent[i : i + 3]))
        print() #extra line for spacing

class SimpleReflexAgent: #agent class
    def __init__(self):
        self.position = 0 #start at 'a' (position 0 in the grid)
        
    def act(self, percept, grid):
        #if the current position is dirty, clean it 
        if percept == 'Dirty':
            grid[self.position] = 'Clean'
            return 'Clean the room'
        else:
            return 'Room is clean'
    
    def move(self):
        #move to the next position
        if self.position < 8:
            self.position += 1
        return self.position
    
    
def run_agent(agent, environment, steps): #simulation function
    for step in range(steps):
        percept = environment.get_percept(agent.position)
        action = agent.act(percept, environment.grid)
        print(f"Step {step + 1}: Position {agent.position} -> Percept - {percept}, Action - {action}")
        environment.display_grid(agent.position) #display the grid state with agent position
        
        if percept == 'Dirty':
            environment.clean_room(agent.position)
        
        agent.move()

#create instances of agent & environment
agent = SimpleReflexAgent()
environment = Environment()

#run the agent in the environment for 9 steps (to cover 3x3 grid)
run_agent(agent, environment, 9)
```

```python
import random

class System:
    def __init__(self):
        self.components = ['V' if random.random() > 0.5 else 'S' for _ in range(9)]
    
    def get_state(self, index):
        return self.components[index]
    
    def patch_component(self, index):
        self.components[index] = 'S'
    
    def display_system(self):
        #display the current state of the system with components A to I
        print("\nCurrent system state:")
        for i in range(0, 9, 3):
            print(" | ".join(self.components[i:i+3]))
        print()  #extra line for spacing
        
class SecurityAgent:
    def __init__(self):
        self.patching_list = []
        
    def scan(self, system):
        #scan system and log vulnerable components
        self.patching_list = []
        for i in range(9):
            if system.get_state(i) == 'V':
                print(f"Component {chr(65 + i)} is vulnerable. added to patch list.")
                self.patching_list.append(i)
            else:
                print(f"Component {chr(65 + i)} is safe.")
                
    def patch(self, system):
        #patch all components in the patching list
        for i in self.patching_list:
            print(f"Patching Component {chr(65 + i)}. now safe.")
            system.patch_component(i)

def run_simulation(agent, system):
    #system check initially
    print("Initial System Check:")
    system.display_system()
    
    #scan system
    print("System Scan:")
    agent.scan(system)
    
    #patching vulnerabilities
    print("\nPatching Vulnerabilities:")
    agent.patch(system)
    
    #final system check
    print("\nFinal System Check:")
    system.display_system()


system = System()
agent = SecurityAgent()

run_simulation(agent, system)
```

```python
import random

class Server:
    def __init__(self, name):
        self.name = name
        # Randomly assign each server a load state: "Underloaded", "Balanced", or "Overloaded"
        self.state = random.choice(["Underloaded", "Balanced", "Overloaded"])

    def __str__(self):
        return f"{self.name}: {self.state}"

class LoadBalancerAgent:
    def __init__(self):
        self.moved_tasks = 0

    def scan_and_balance(self, system):
        underloaded_servers = []
        overloaded_servers = []

        # Scan the system for underloaded and overloaded servers
        for server in system.servers:
            if server.state == "Underloaded":
                underloaded_servers.append(server)
            elif server.state == "Overloaded":
                overloaded_servers.append(server)

        # Redistribute tasks: move tasks from overloaded to underloaded servers
        for overloaded in overloaded_servers:
            if underloaded_servers:
                # Move a task from an overloaded server to an underloaded server
                underloaded = underloaded_servers.pop()
                print(f"Moving tasks from {overloaded.name} to {underloaded.name}")
                overloaded.state = "Balanced"  # Overloaded becomes Balanced
                underloaded.state = "Balanced"  # Underloaded becomes Balanced
                self.moved_tasks += 1

        # Update states for any servers that remained underloaded or overloaded
        if self.moved_tasks == 0:
            print("No tasks were moved, the system is already balanced!")

class System:
    def __init__(self):
        self.servers = [Server(f"Server {i+1}") for i in range(5)]

    def display_system(self):
        print("\nCurrent System Load Status:")
        for server in self.servers:
            print(server)

def run_simulation(agent, system):
    print("Initial System Check:")
    system.display_system()

    # Load Balancer scans and balances the system
    print("\nLoad Balancer is balancing the load...\n")
    agent.scan_and_balance(system)

    # Final system check after balancing
    print("\nFinal System Check:")
    system.display_system()


# Create the system and load balancer agent
system = System()
agent = LoadBalancerAgent()

# Run the simulation
run_simulation(agent, system)
```
